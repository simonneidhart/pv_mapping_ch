```python
import pandas as pd
```

```python
# There are some NaN rows in the data, which are dropped.
pv_meters = pd.read_excel("../../data/Liste_PV_available.xlsx").dropna(how="all")
```

```python
# Renaming columns to conform to data model.
pv_meters["utility"] = pv_meters["Utility"]
pv_meters["municipality"] = pv_meters["Ort"]
pv_meters["address"] = pv_meters["Adress"]

pv_meters["installed_capacity_kw"] = pv_meters["installed capacity"]
pv_meters["orientation_deg"] = pv_meters["orientation 1"]
pv_meters["slope_deg"] = pv_meters["slope"]

# Resetting index because we will use it as a unique identifier.
pv_meters.reset_index(inplace=True)
pv_meters["id"] = pv_meters.index
```

```python
data_clean = pv_meters[
    [
        "id",
        "lat",
        "lon",
        "address",
        "municipality",
        "utility",
        "installed_capacity_kw",
        "slope_deg",
        "orientation_deg",
    ]
]
```
